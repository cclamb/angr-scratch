```python
import angr

h = lambda x: format(x, '#16x')

p = angr.Project('./fauxware')

RANGE = 10

def project(p):
    print '############'
    print 'Executable name: %s' % (p.filename)
    print 'PIC: %s XSTACK: %s' % (p.loader.main_object.pic, p.loader.main_object.execstack)
    print p.arch.name, p.arch.memory_endness, p.arch.bytes, p.arch.bits
    print 'Entry Point: %s' % (h(p.entry))
    print '############'
    print ''

def block(p, a):
    symbol = p.loader.find_symbol(a)
    print '-> Block starting at %s' % (h(a))
    print '-> Symbol: %s' % (symbol)
    print '-> START'
    b = p.factory.block(a)
    b.pp()
    print '-> END'
    print ''
    return symbol.name if symbol is not None else 'None'
    
def step(p):
    state = p.factory.entry_state()
    sm = p.factory.simgr(state)
    while True:
        a = sm.active[0]
        symbol = block(p, a.addr)
        sm.step()
        if symbol == 'main':
            break
    
    
project(p)
step(p)
```
