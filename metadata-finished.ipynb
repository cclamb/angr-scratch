```python
import angr

h = lambda x: format(x, '#16x')

p = angr.Project('./fauxware')

def project(p):
    print '############'
    print 'Executable name: %s' % (p.filename)
    print 'PIC: %s XSTACK: %s' % (p.loader.main_object.pic, p.loader.main_object.execstack)
    print p.arch.name, p.arch.memory_endness, p.arch.bytes, p.arch.bits
    print 'Entry Point: %s' % (h(p.entry))
    print '############'
    print ''
    
def loaded_objects(p):
    print '# OBJECTS ##'
    for o in p.loader.all_objects:
        print o
    print '############'
    print ''
        
def segments(p):
    print '# SEGMENTS #'
    for seg in p.loader.main_object.segments:
        print seg
    print '############'
    print ''
        
def sections(p):
    print '# SECTIONS #'
    for sec in p.loader.main_object.sections:
        print sec
    print '############'
    print ''
    
def plt(p):
    print '# PLT ######'
    for plt in p.loader.main_object.plt:
        print plt
    print '############'
    print ''
    
def imports(imports):
    print '# IMPORTS ##'
    for imprt in p.loader.main_object.imports:
        print imprt
    print '############'
    print ''
    
    
project(p)
loaded_objects(p)
segments(p)
sections(p)
plt(p)
imports(p)
```
