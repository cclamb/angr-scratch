```python
import angr
```

```python
class Printer(object):
    
    _t = '[+]'
    _f = '[-]'
    _e = '[=]'
    
    def __init__(self, b):
        self._b = b
        
    def _print_imports(self, imports, name='main'):
        print '%s Main imported symbols for %s:' % (Printer._e, name)
        for imprt in imports:
            sym = imports[imprt]
            print '%s\tName: %s' % (Printer._e, imprt)
            # print sym
            
    def main(self):
        print '%s Entry point: 0x%0x' % (Printer._e, self._b.entry)
        print '%s Range  [0x%0x -> 0x%0x]' % (Printer._e, self._b.loader.min_addr(), self._b.loader.max_addr())
        print '%s Name: %s' % (Printer._e, self._b.filename)

    def libraries(self):
        l = self._b.loader
        print '%s Loader: %s' % (Printer._e, l)
        for obj in l.shared_objects:
            print '%s\tso: %s' % (Printer._e, obj)
        print '%s Memory: %s' % (Printer._e, l.main_bin)
        print '%s Object @ 0x%0x: %s' % (Printer._e, l.min_addr(), l.addr_belongs_to_object(l.min_addr()))
        print '%s main(.) addr in GOT: 0x%0x'  % (Printer._e, l.find_symbol_got_entry('__libc_start_main'))
        
    def main_details(self):
        m = self._b.loader.main_bin
        print '%s Main binary: %s' % (Printer._e, m)
        print '%s Main memory: %s' % (Printer._e, m.memory)
        self._print_imports(m.imports)
        
    def library_details(self):
        for so_name in self._b.loader.shared_objects:
            so = self._b.loader.shared_objects[so_name]
            self._print_imports(so.imports, so_name)
            
             
b = angr.Project('/bin/true')
printer = Printer(b)
printer.main()
printer.libraries()
printer.main_details()
printer.library_details()
```
